```python
#### Graficar la cantidad de películas producidas cada año, de 2006 a 2016

## Tu código aquí

#Determine the amount of movies that were produced each year, from 2006-2016

dfMovies6_16 = dfMovies[(dfMovies['title_year'] >= 2006) & (dfMovies['title_year'] <= 2016)]

movie_yearly_count = dfMovies6_16['title_year'].value_counts().sort_index().plot(kind='bar', color='r', alpha=0.5, grid=False, rot=45)
movie_yearly_count.set_xlabel('Year')
movie_yearly_count.set_ylabel('Movie Count')
movie_yearly_count.set_title('Movie Count by Year')

plt.show()


dfMovies6_16['title_year'].value_counts().sort_index()

```

```python
#### Obtener top 20 ganancias totales por director

## Tu código aquí


#dfMovies['director_name'].value_counts()

dfMovies['total_revenue'] = (dfMovies['gross'] - dfMovies['budget']) / 1000000
dfMovies.head()

total_director_revenue = dfMovies[['director_name','total_revenue']].groupby('director_name').sum().reset_index()

total_director_revenue = total_director_revenue.sort_values(by=['total_revenue'], ascending = [False])

total_director_revenue.head(20)
```
